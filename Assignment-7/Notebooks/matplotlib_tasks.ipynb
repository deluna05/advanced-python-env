```python
import matplotlib.pyplot as plt
```

```python
names = ["Amina", "Dias", "Aruzhan", "Timur", "Aiho"]
scores = [88, 75, 91, 84, 95]
```

```python
plt.plot(names, scores, marker="o")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores (Line Plot)")
plt.show()
```

```python
plt.bar(names, scores)
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores (Bar Chart)")
plt.show()
```
